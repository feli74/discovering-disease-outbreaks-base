```python
import geonamescache as gnc
from unidecode import unidecode
import re
import pandas as pd


gc = gnc.GeonamesCache()
countries = gc.get_countries()
cities = gc.get_cities()

city_lookup = {}
for key, value in cities.items():
    name = unidecode(value['name'])
    city_lookup[name.lower()] = value

regex_pattern = "\\b({0})\\b".format("|".join(city_lookup.keys()))
regex = re.compile(regex_pattern, flags=re.IGNORECASE)

with open('data/headlines.txt') as f:
    lines = f.read().splitlines()
    
```

```python
found_cities = []

for line in lines:
    match = regex.match(line)
    if None != match:
        groups = match.groups()
        groups = [i for i in groups if i]
        if len(groups) > 0:
            city_name = groups[0]
            city_data = city_lookup[city_name.lower()]
            data = {}
            data['headline'] = line
            data['city'] = city_data['name']
            data['country'] = city_data['countrycode']
            found_cities.append(data)

#print(*found_cities, sep = "\n") 

```

```python
zipped = { 'headline': [], 'city': [], 'country': []}
for each in found_cities:
    zipped['headline'].append(each['headline'])
    zipped['city'].append(each['city'])
    zipped['country'].append(each['country'])
    
df = pd.DataFrame(zipped, columns=['headline', 'city', 'country'])
print(df)
df.to_pickle("headline_city_country.pkl")
```
